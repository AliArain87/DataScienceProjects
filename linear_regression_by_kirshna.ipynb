```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
```

```python
df = load_boston()
df
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head(10)
```

```python
## independent and dependent features
X = dataset
y = df.target
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.3 , random_state=42)

```

```python
X_train
```

```python
X_test
```

```python
## Standarizing the dataset
## we used standardScaler for standardized our dataset
## Remember we use fit_transform for input x_train
## and transform() for output X_test

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
## hyper-parameter tune??
## cross validation?? we divide data into parts and train the parts of data and find the accuracy of each data and we find the mean accuracy
from sklearn.linear_model import LinearRegression
## cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
mean_square_error = cross_val_score(regression ,X_train , y_train , scoring='neg_mean_squared_error',cv=10 )

```

```python
np.mean(mean_square_error)
```

```python
## prediction on test data
reg_pred = regression.predict(X_test)

```

```python
reg_pred
```

```python
sns.displot(reg_pred - y_test, kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_pred , y_test)
```
